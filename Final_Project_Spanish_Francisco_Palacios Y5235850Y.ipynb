## Tarea 2b: Limpieza de datos (Python)

```python
## Comprueba si hay valores nulos y muestra el recuento en orden ascendente.
df.isnull().sum().sort_values()
```

```python
for col in df.columns:
    if df[col].dtype == 'O':
        print("Rellenando valores nulos en la columna categórica:", col)
        df[col].fillna(value=df[col].mode(), inplace=True)
    else:
        print("Rellenando valores nulos en la columna numérica:", col)
        df[col].fillna(value=df[col].median(), inplace=True)
```

```python
df.isnull().sum().sort_values()
```

```python
## Comprueba si hay valores duplicados y elimínalos.

df.duplicated().sum()

```

```python
#y elimínalos.
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
top10barrios_price=df4.groupby(by="neighbourhood")["price"].mean().nlargest(10)
df5=pd.DataFrame(top10barrios_price)
df6=df5.transpose()
```

```python
fig,ax=plt.subplots(1,1)
fig.set_size_inches(11.7,8.27)

vis1= sns.barplot(df6,orient="h")

plt.ylabel("Neighbourhood", fontsize=15,color="black")#etiqueta leyenda 
plt.xlabel("Price", fontsize=15,color="Green") #etiqueta leyenda 
ax.set_title("Top 10 barrios más caros",size = 20)


plt.show()
```

```python
#¿Cuál es el barrio más barato? Lighthouse Hill
top10barrios_price=df4.groupby(by="neighbourhood")["price"].mean().nsmallest(1)
df5=pd.DataFrame(top10barrios_price)
df6=df5
df6
```

```python
#Enumere los barrios que ofrecen alquileres a corto plazo de menos de 10 días. Ilustrar con un gráfico de barras
df.minimum_nights=df.minimum_nights.astype("object")
menos_10_días=df[["neighbourhood","minimum_nights"]]
filtro_menos_10_días=menos_10_días.minimum_nights.isin([9,8,7,6,5,4,3,2,1])
#filtro_menos_10_días
```

```python
df5aa=menos_10_días[filtro_menos_10_días]
df5aa_1=df5aa.groupby(by="neighbourhood")["minimum_nights"].mean()
df5aaa=pd.DataFrame(df5aa_1)
df5aaa=df5aaa.minimum_nights.astype("int")
freq2=df5aaa.value_counts()
freq3=pd.DataFrame(freq2)
freq4=freq3.transpose()
```

```python
sns.set_theme()
fig, ax = plt.subplots(figsize = (9, 3.5))
vis2= sns.barplot(freq4)
vis2.figure.suptitle("Numero de barrios que ofrecen alquileres a corto plazo", fontsize=18,
    fontdict={"weight": "bold"})

ax.set(ylabel=" Nº barrios",
       xlabel="Minimum nights")

plt.rcParams["figure.figsize"]=4,8
```

## Tarea 5b: Visualización de datos (Cualquier herramienta)

```python
#Crea un gráfico de dispersión para ilustrar la relación entre la tarifa de limpieza y el precio de la habitación
#y escribe el tipo de correlación que observas, si la hay.

df.service_fee=df.service_fee.astype("int")


fig,ax=plt.subplots(1,1)
sns.set_theme()
sns.scatterplot(x=df.service_fee, y=df.price)


plt.xlabel('Tarifa de limpieza')

plt.ylabel('Precio de la habitación')

ax.set_title("Relación entre tarifa de limpieza y el precio de la habitación",size = 15)

plt.show()
```

```python
correlación_coefi= df['service_fee'].corr(df['price'])
print(correlación_coefi)

#siendo una correlacion muy cercana a 1, asumimos que ambas variables se mueven juntas. mientras mayor es el 
#precio de la habitación, mayor es la tarifa de limpieza. 
```

```python
#Crea un gráfico de líneas para mostrar la cantidad total de lugares disponibles al año.

lugares_por_año = df['construction_year'].value_counts().sort_index()
lugares_por_año
```
